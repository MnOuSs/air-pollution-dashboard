### First Data Check

```python
import pandas as pd

df = pd.read_csv("../data/air_pollution.csv", sep=";")

df.head()
```

```python
df.columns
```

### Cleaning the Dataset

```python
# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
# Keeping only relevant columns
df = df[['city', 'country_label', 'coordinates', 'pollutant', 'value', 'last_updated']]
df.head()
```

```python
# Filter for PM2.5 pollutant
df = df[df['pollutant'] == 'PM2.5']
df.head()
```

```python
# Rename columns
df = df.rename(columns={
    'country_label': 'country',
    'value': 'pm25'
})

df.head()
```

```python
# Extract latitude and longitude from coordinates
df[['latitude', 'longitude']] = df['coordinates'].str.split(',', expand=True).astype(float)
df.head()
```

```python
# Drop the original coordinates column
df = df.drop(columns=['coordinates'])
df.head()
```

```python
# Replacing missing cities with 'Unknown Location'
df['city'] = df['city'].fillna('Unknown Location')
df.head()
```

```python
# Removing missing values
df = df.dropna(subset=['latitude', 'longitude', 'pm25'])
df.head()
```

```python
# Removing outliers in PM values
df = df[(df['pm25']> 0) & (df['pm25'] < 500)]
df.head()
```

## Final Check

```python
df.describe()
```

## Basic Map

```python
import plotly.express as px

fig = px.scatter_geo(df, lat='latitude', lon='longitude', color='pm25',
                     size='pm25', hover_name='country', projection='natural earth',
                        title='Global PM2.5 Pollution Levels')

fig.show()
```

## Improved Map

```python
fig = px.scatter_geo(df, lat='latitude', lon='longitude', color='pm25',
                     size='pm25', hover_name='country', projection='natural earth',
                        title='Global PM2.5 Pollution Levels')

fig.update_layout(geo=dict(showland=True, landcolor='rgb(243, 243, 243)', showcountries=True),
                  margin={"r":0,"t":50,"l":0,"b":0})

fig.show()
```

### First Insights

```python
# Top polluted countries
top_countries = df.groupby('country')['pm25'].mean().sort_values(ascending=False).head(10)
print(top_countries)
```

```python
# Plotting the result
fig = px.bar(top_countries, x=top_countries.values, y=top_countries.index, orientation='h',
                 title='Top 10 Most Polluted Countries by Average PM2.5 Levels',
                 labels={'x': 'Average PM2.5 Level', 'y': 'Country'})

fig.show()
```
